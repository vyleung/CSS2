# Block 1

```python
# imported the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
anscombe_data = sns.load_dataset('anscombe')
```

# Block 2

```python
# assigned each dataset to a variable
dataset1 = anscombe_data[anscombe_data['dataset'] == 'I'] # dataset I
dataset2 = anscombe_data[anscombe_data['dataset'] == 'II'] # dataset II
dataset3 = anscombe_data[anscombe_data['dataset'] == 'III'] # dataset III
dataset4 = anscombe_data[anscombe_data['dataset'] == 'IV'] # dataset IV

# created a 2x2 subplot
anscombe_fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(15,10))

# used a nested for loop to go through the 4 datasets
for i in np.arange(0,axes.shape[0]):
  for j in np.arange(0,axes.shape[1]):
    if i == 0 and j == 0: # top left subplot
      axes[i,j].plot(dataset1['x'],dataset1['y'],linestyle='',marker='o', color='blue')
      # set x and y-axis labels
      axes[i,j].set_xlabel('$x_1$')
      axes[i,j].set_ylabel('$y_1$')
      # set subplot title to 'Dataset I'
      axes[i,j].set_title('Dataset I')
      # put text inside the plot w/ the mean and std dev for both x and y variables
      axes[i,j].annotate('mean of x: '+str(round(np.mean(dataset1['x']),2)), xy=(15,5))
      axes[i,j].annotate('std dev of x: '+str(round(np.std(dataset1['x']),2)), xy=(15,4))
      axes[i,j].annotate('mean of y: '+str(round(np.mean(dataset1['y']),2)), xy=(15,3))
      axes[i,j].annotate('std dev of y: '+str(round(np.std(dataset1['y']),2)), xy=(15,2))
    
    elif i == 0 and j != 0: # top right subplot
      axes[i,j].plot(dataset2['x'],dataset2['y'],linestyle='',marker='^', color='green')
      # set x and y-axis labels
      axes[i,j].set_xlabel('$x_2$')
      axes[i,j].set_ylabel('$y_2$')
      # set subplot title to 'Dataset II'
      axes[i,j].set_title('Dataset II')
      # put text inside the plot w/ the mean and std dev for both x and y variables
      axes[i,j].annotate('mean of x: '+str(round(np.mean(dataset2['x']),2)), xy=(15,5))
      axes[i,j].annotate('std dev of x: '+str(round(np.std(dataset2['x']),2)), xy=(15,4))
      axes[i,j].annotate('mean of y: '+str(round(np.mean(dataset2['y']),2)), xy=(15,3))
      axes[i,j].annotate('std dev of y: '+str(round(np.std(dataset2['y']),2)), xy=(15,2))

    elif i !=0 and j == 0: # bottom left subplot
      axes[i,j].plot(dataset3['x'],dataset3['y'],linestyle='',marker='*', color='red')
      # set x and y-axis labels
      axes[i,j].set_xlabel('$x_3$')
      axes[i,j].set_ylabel('$y_3$')
      # set subplot title to 'Dataset III'
      axes[i,j].set_title('Dataset III')
      # put text inside the plot w/ the mean and std dev for both x and y variables
      axes[i,j].annotate('mean of x: '+str(round(np.mean(dataset3['x']),2)), xy=(15,5))
      axes[i,j].annotate('std dev of x: '+str(round(np.std(dataset3['x']),2)), xy=(15,4))
      axes[i,j].annotate('mean of y: '+str(round(np.mean(dataset3['y']),2)), xy=(15,3))
      axes[i,j].annotate('std dev of y: '+str(round(np.std(dataset3['y']),2)), xy=(15,2))

    else: # bottom right subplot
      axes[i,j].plot(dataset4['x'],dataset4['y'],linestyle='',marker='s', color='orange')
      # set x and y-axis labels
      axes[i,j].set_xlabel('$x_4$')
      axes[i,j].set_ylabel('$y_4$')
      # set subplot title to 'Dataset IV'
      axes[i,j].set_title('Dataset IV')
      # put text inside the plot w/ the mean and std dev for both x and y variables
      axes[i,j].annotate('mean of x: '+str(round(np.mean(dataset4['x']),2)), xy=(15,5))
      axes[i,j].annotate('std dev of x: '+str(round(np.std(dataset4['x']),2)), xy=(15,4))
      axes[i,j].annotate('mean of y: '+str(round(np.mean(dataset4['y']),2)), xy=(15,3))
      axes[i,j].annotate('std dev of y: '+str(round(np.std(dataset4['y']),2)), xy=(15,2))

    # set all subplots' x axis limit and tick marks to the greatest x value in the dataset+2 w/ all subplots' x axis incremented by 2
    axes[i,j].set_xlim([0,anscombe_data['x'].max() + 2])
    axes[i,j].set_xticks(np.arange(0,anscombe_data['x'].max()+2, 2))
    # set all subplots' y axis limit and tick marks to the greatest y value in the dataset+2 w/ all subplots' y axis incremented by 2
    axes[i,j].set_ylim([0,anscombe_data['y'].max() + 2])
    axes[i,j].set_yticks(np.arange(0,anscombe_data['y'].max()+2, 2))

plt.tight_layout()
```
